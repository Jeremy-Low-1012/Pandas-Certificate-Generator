## 5. Creating Certificate logic

```python
def certificate_generator(name,courseName,courseLevel,date):
  #  store the generated file name inside a variable "pdf_file_name"
  # For example, when using this function (to yeekean-python-begginer-pdf )
  pdf_file_name="/content/drive/MyDrive/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificates/" + name + "-" + courseName + "-" + courseLevel + ".pdf"

  # canvas =  blank screen (A4 landscape size)

  # we are storing this canvas inside this .pdf

  # basically we have a blank landscape A4 virtual paper in our .pdf file


  c = canvas.Canvas(pdf_file_name,pagesize=landscape(A4))
  # the canvas' drawImage() 5 inputs
  # 1st input -- image --> what are you drawing on this empty canvas?? --> certficate template
  # 2nd input --> x axis (coordinate 0)
  # 3rd input --> y axis (coordinate 0)
  # 4th input --> width --> A4[1] --> (using the A4 variable we imported)
  # 5th input --> height --> A4[0] --> (using the A4 variable we imported)
  c.drawImage("/content/drive/MyDrive/1.2 Python Certificate Generator Project (STARTING TEMPLATE)/certificate_template.jpg",0,0,width=A4[1],height=A4[0])

  center_x=c._pagesize[0]/2


  center_y=c._pagesize[1]/2

  c.setFont("Lora-Bold",30)
  c.drawCentredString(center_x,center_y-46,name)

  c.setFont("Lora-Bold",28)
  c.drawCentredString(center_x,center_y-105,courseName+" - " + courseLevel)

  c.setFont("Lora-Bold",30)
  c.drawCentredString(center_x,center_y-46,name)

  c.setFont("Lora-Bold",17)
  c.drawCentredString(center_x+190,center_y-160,date)

  # 4.Cert Id
  cert_id="Cert ID"+ str(int(pd.Timestamp.now().timestamp()))
  c.setFont("Lora-Bold",12)
  c.drawCentredString(center_x+ 266,center_y - 230,cert_id.upper())




  c.save()
```
